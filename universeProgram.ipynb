```python
import numpy as np
import matplotlib.pyplot as plt
import math
import random
```

```python
def fileInitialization(fileName):
  with open(fileName, "w") as f:
    f.write("U = { e, ")
```

```python
def closeFile(fileName):
  with open(fileName, "a") as f:
    f.write("}")
```

```python
def getN():
  global n
  opc = input("Hello! Do you want to insert a value or have the program generate it automatically? I (I insert the value) / P (the program generates it)")
  if opc == 'I':
    n = int(input("Enter the maximum length number: "))
  if opc == 'P':
    n = random.randint(0, 1000)
```

```python
def countStrings():
  global totalStrings
  global xStrings
  totalStrings += 1

  xStrings.append(totalStrings)
```

```python
def countOnes(totalOnes):
  global yOnes

  yOnes.append(totalOnes)
```

```python
def write(string):

  countStrings()

  global totalOnes
  totalOnes = 0
  for character in string:
    if character == '1':
      totalOnes += 1
    else:
      continue
  countOnes(totalOnes)

  with open("universe.txt", "a") as f:
    f.write(string + ", " )
```

```python
def generar_cadenas(n, alfabeto):
    if n == 0:
        return [""]
    elif n == 1:
        return alfabeto
    else:
        cadenas_longitud_n_minus_1 = generar_cadenas(n - 1, alfabeto)
        cadenas_longitud_n = []
        for cadena_n_minus_1 in cadenas_longitud_n_minus_1:
            write(cadena_n_minus_1)
            for caracter in alfabeto:
                cadenas_longitud_n.append(cadena_n_minus_1 + caracter)
        return cadenas_longitud_n
```

```python
def doGraph(x, y, xLabel, yLabel, title, widthSize, heightSize):

  plt.figure(figsize=(widthSize, heightSize))

  plt.plot(x, y, marker='o', linestyle='-')

  plt.xlabel(xLabel)
  plt.ylabel(yLabel)
  plt.title(title)

  plt.grid(True)
  plt.show()
```

```python
def logarithm(aList):
  logarithm_vector=[]
  for x in aList:
    if x > 0:
        logarithm_vector.append(math.log(x))
    else:
        logarithm_vector.append(float('nan'))
  return logarithm_vector
```

```python
while True:

  n=0
  xStrings = [1]
  yOnes = [0]
  totalOnes = 0
  totalStrings = 1 # Add empty string
  fileName = "universe.txt"
  alphabet = ["0", "1"]

  getN()

  fileInitialization(fileName)

  cadenas_longitud_n = generar_cadenas(n, alphabet)

  closeFile(fileName)

  response = input("Do you want to repeat the program with another value of the maximum length? (Yes/No): ")
  if response.lower() != "yes":
        break
```

```python
doGraph(xStrings, yOnes, "Strings", "Count Ones", "Universe of strings", 30, 10)
```

```python
logarithm_xStrings = logarithm(xStrings)
logarithm_yOnes = logarithm(yOnes)
```

```python
doGraph(logarithm_xStrings, logarithm_yOnes, "Strings", "Count Ones", "Universe of strings (logarithm values)", 30, 10)
```
